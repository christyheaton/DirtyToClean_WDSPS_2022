This is a small dataset used to demonstrate some common ways environmental data can be analyzed and visualized.

```python
# load data
results_path = r'../input/EIMResults_2022Mar17_17883.csv'
results_df = pd.read_csv(results_path)
```

### Preliminary Data Cleanup and Review

Groundwater is the water found underground in the cracks and spaces in soil, sand and rock. It is stored in and moves slowly through geologic formations of soil, sand and rocks called aquifers.

Groundwater monitoring wells are used for determining groundwater quality over time.
Groundwater can also be sampled as a single event along with soil sampling.

```python
# keep only groundwater data
gw_df = results_df[results_df['Sample_Source']=='Groundwater']
```

```python
# keep only relevant columns
gw_df = gw_df[['Location_ID',
               'Field_Collection_Start_Date',
               'Result_Parameter_Name',
               'Result_Value',
               'Result_Value_Units',
               'Result_Data_Qualifier',
               'Calculated_Latitude_Decimal_Degrees_NAD83HARN',
               'Calculated_Longitude_Decimal_Degrees_NAD83HARN',
               'Lower_Depth_In_Feet']]

# rename coordinate columns to be shorter
gw_df.rename({'Calculated_Latitude_Decimal_Degrees_NAD83HARN':'Lat',
              'Calculated_Longitude_Decimal_Degrees_NAD83HARN':'Long'},axis=1,inplace=True)
```

```python
gw_df.head(3)
```

```python
# use the coordinate columns to convert to a GeoPandas GeoDataFrame
gw_gdf = gpd.GeoDataFrame(gw_df, geometry=gpd.points_from_xy(gw_df.Long, gw_df.Lat), crs=4326)

# geopandas does not support datetime columns
gw_gdf['Date_str'] = pd.to_datetime(gw_gdf['Field_Collection_Start_Date']).dt.strftime('%Y-%m-%d')

# retain only data from 2014 or later
gw_gdf = gw_gdf[gw_gdf['Date_str']>'2014-1-1']
```

```python
# note the geometry column and that there is one record per result
gw_gdf.head(3)
```

```python
# create a geodataframe with only location IDs and geometry for quick visual inspection
locs = gw_gdf[['Location_ID','geometry']].drop_duplicates()
```

```python
locs.head(3)
```

```python
# load in important areas
aoc = gpd.read_file(r'../input/aoc.geojson')
site_area = gpd.read_file(r'../input/site_area.geojson')
```

```python
# to display multiple layers on an interactive map, set the bottom layer to a variable and 
# call that variable in future layers
m = site_area.explore(tiles='Stamen Toner', color='teal')
aoc.explore(m=m, color='orange')
locs.explore(m=m, color='blue', marker_kwds=dict(radius=3))
```

### Is there contamination at the site?

```python
# define chemicals of concern
COCs = ['Cis-1,2-Dichloroethene',
        'Vinyl Chloride',
        'Trichloroethene',
        'Trans-1,2-Dichloroethene',
        'Tetrachloroethene']
```

```python
# keep only result records that are related to a chemical of concern
gw_gdf = gw_gdf[gw_gdf['Result_Parameter_Name'].isin(COCs)]
```

```python
# chemical names are long, so create abbreviations to use going forward
def map_values(row, values_dict):
    return values_dict[row]

Abbr = {'Cis-1,2-Dichloroethene': 'cDCE',
        'Vinyl Chloride':'VC',
        'Trichloroethene':'TCE',
        'Trans-1,2-Dichloroethene':'tDCE',
        'Tetrachloroethene':'PCE'}

gw_gdf['Abbr'] = gw_gdf['Result_Parameter_Name'].apply(map_values, args = (Abbr,))
```

```python
# verify data is in consistent units
gw_gdf.Result_Value_Units.unique()
```

```python
# chemicals are not dangerous unless they exceed a certain concentration
# these values vary depending on location, expected contact with humans and animals, and other factors

# define groundwater cleanup levels in ug/L
CULs = {'cDCE':70,
        'VC':0.20,
        'TCE':5.0,
        'tDCE':100,
        'PCE':5.0}
```

```python
# labs can only detect data down to a certain value depending on their equipment and technology
# if they can't detect a chemical they will provide the minimum detectable value, which is approximate
# this number can very between labs and can change as new technology gets developed
# sometimes these minimum detectable values exceed cleanup levels, so in this example we want to 
# make sure that they never do

gw_gdf['Result'] = gw_gdf['Result_Value']
gw_gdf.loc[gw_gdf['Result_Data_Qualifier'] == 'U', 'Result'] = -1
```

```python
# calculate the exceedance factor of each result
for index, row in gw_gdf.iterrows():
    cul = CULs[row['Abbr']]
    ex_fac=0
    if row['Result'] > cul:
        ex_fac = 1
    if row['Result'] > cul*2:
        ex_fac = 2
    if row['Result'] > cul*10:
        ex_fac = 10
    gw_gdf.loc[index, 'Exceedance_Factor'] = ex_fac
```

```python
gw_gdf.head(3)
```

```python
# check the maximum exceedance factor to see if there is contamination
gw_gdf.Exceedance_Factor.max()
```

### What is the extent of contamination by COC?

```python
# create a new dataframe that only has the maximum result by chemical for each location
gw_gdf_max = gw_gdf.loc[gw_gdf.groupby(['Location_ID','Abbr',])['Result'].idxmax()] 
```

```python
# we need to handle each chemical separately, so to prepare, create a dictionary where keys are
# the chemical names and values are the geodataframes

GeoDataFrameDictAnalytes = {elem : pd.DataFrame for elem in list(Abbr.values())}
for key in GeoDataFrameDictAnalytes.keys():
    GeoDataFrameDictAnalytes[key] = gw_gdf_max[:][gw_gdf_max.Abbr == key]
```

```python
GeoDataFrameDictAnalytes.keys()
```

```python
GeoDataFrameDictAnalytes.values()
```

```python
# create an interactive map of the results, with a different layer for each chemical
m = folium.Map(location=[47.76, -122.207], zoom_start=17, tiles='Stamen Toner')
site_area.explore(m=m, color='teal', name='Site Area')
aoc.explore(m=m, color='orange', name='Area of Concern')

for coc in GeoDataFrameDictAnalytes.keys():   
    gdf = GeoDataFrameDictAnalytes[coc] 
    gdf.explore(m=m,
            name=gdf['Abbr'].iloc[0],
            cmap='autumn_r',
            show=False,
            column='Exceedance_Factor',
            scheme='userdefined',classification_kwds={'bins':[0,1,2,10]},
            marker_kwds=dict(radius=6),
            legend_kwds=dict(colorbar=False, labels=['Meets CUL or ND','>1x CUL','>2xCUL','>10xCUL']),
            tooltip=['Location_ID','Date_str','Abbr','Result_Value', 'Result_Data_Qualifier','Exceedance_Factor']
            )
    
folium.LayerControl(collapsed=True, autoZIndex=True).add_to(m)
```

```python
# save the map and send to someone else
m.save('../output/max_by_chem.html')
```

### How deep does it go?

Depending on the hydrology of the site it can make sense to determine aquifer zones and then group data into these zones.

```python
# create geodataframes for the different aquifer zones - shallow is < 25 feet deep, deep is >= 25 feet deep
shallow_gw =  gw_gdf[gw_gdf['Lower_Depth_In_Feet']<25]
deep_gw = gw_gdf[gw_gdf['Lower_Depth_In_Feet']>=25]
```

```python
locs.head(3)
```

```python
def pivgeo(gdf):
    '''pivot to determine max exceedance factor for each chemical at each location and convert back into geodataframe'''
    gw_piv = pd.pivot_table(gdf, values='Exceedance_Factor', index=['Location_ID'], columns=['Abbr'], aggfunc=np.max)
    gw_piv.reset_index(level=0, inplace=True)
    gw_piv = gw_piv.fillna(0)
    # determine the maximum exceedance factor for each chemical
    for index, row in gw_piv.iterrows():
        max_ex_fac = max([row['PCE'],row['TCE'],row['VC'],row['cDCE'],row['tDCE']])
        gw_piv.loc[index, 'Maximum Exceedance Factor'] = max_ex_fac
    
    # merge the locs geodataframe to the pivot table to get the geometry back (it was lost in the pivot step)
    gw_piv_merge = pd.merge(gw_piv, locs, how='left',on='Location_ID')
    # redfine the geodataframe using the geometry column
    gw_piv_gdf = gpd.GeoDataFrame(gw_piv_merge, geometry='geometry', crs=4326)
    return gw_piv_gdf
```

```python
# run the function on the shallow and deep aquifer zone geodataframes
shallow_piv_gdf = pivgeo(shallow_gw)
deep_piv_gdf = pivgeo(deep_gw)
```

```python
shallow_piv_gdf.head()
```

```python
deep_piv_gdf.head()
```

```python
# map the results to see the difference in exceedances between the two aquifer zones
m = folium.Map(location=[47.76, -122.207], zoom_start=16, tiles='Stamen Toner')
site_area.explore(m=m, color='teal', name='Site Area')
aoc.explore(m=m, color='orange', name='Area of Concern')

shallow_piv_gdf.explore(m=m,
                   name='Shallow',
                   show=False,
                   cmap='autumn_r',
                   column='Maximum Exceedance Factor',
                   scheme='userdefined',classification_kwds={'bins':[0,1,2,10]},
                   marker_kwds=dict(radius=6),
                   legend_kwds=dict(colorbar=False, labels=['Meets CUL or ND','>1x CUL','>2xCUL','>10xCUL']),
                   )

deep_piv_gdf.explore(m=m, 
                   name='Deep',
                   show=False,
                   cmap='autumn_r',
                   column='Maximum Exceedance Factor',
                   scheme='userdefined',classification_kwds={'bins':[0,1,2,10]},
                   marker_kwds=dict(radius=6),
                   legend_kwds=dict(colorbar=False, labels=['Meets CUL or ND','>1x CUL','>2xCUL','>10xCUL']),
                   )

folium.LayerControl(collapsed=True, autoZIndex=True).add_to(m)
```

### Is it changing?

```python
# create a column to determine if a location is inside the area of concern

# use the geopandas sjoin method with the 'intersects' option
gw_gdf_aoc = gpd.sjoin(gw_gdf, aoc, predicate='intersects')

# create an InAOC column in the new geodataframe
gw_gdf_aoc['InAOC'] = 1

# simplify so there is only one row per location
aoc_locs = gw_gdf_aoc.groupby(['Location_ID'])['InAOC'].max()

# merge the simplified geodataframe back to the main geodataframe to get the InAOC column
gw_gdf = gw_gdf.merge(aoc_locs, how = 'outer', on = ['Location_ID'])

# any row without a value in InAOC should be set to 0 (it is not inside the AOC)
gw_gdf['InAOC'] = gw_gdf['InAOC'].fillna(0)
```

```python
# turn the geodataframe into a non-spatial dataframe to prepare for time series charts
gw_df_ex = pd.DataFrame(gw_gdf)

# reset the date column to datetime type
gw_df_ex['Date']= pd.to_datetime(gw_df_ex['Date_str'])
```

```python
# create a dictionary where keys are the analyte names and values are the dataframes
DataFrameDictAnalytes = {elem : pd.DataFrame for elem in list(Abbr.values())}
for key in DataFrameDictAnalytes.keys():
    DataFrameDictAnalytes[key] = gw_df_ex[:][gw_df_ex.Abbr == key]
```

```python
# view the monthly averages of each chemical site-wide and in the area of concern
for coc in DataFrameDictAnalytes.keys():   
    
    # define site dataframe and aoc data frame
    df = DataFrameDictAnalytes[coc]
    df_aoc = df[df['InAOC']==1]
    
    # group by month
    df = df.groupby(pd.PeriodIndex(df['Date'], freq='M'))['Result_Value'].mean().reset_index()
    df_aoc = df_aoc.groupby(pd.PeriodIndex(df_aoc['Date'], freq='M'))['Result_Value'].mean().reset_index()
    
    # show both dataframes on the plot
    ax = df.plot(x='Date', y='Result_Value', figsize=(15, 5), ls='-', color='teal',label='Site-wide',linewidth=2.5)
    df_aoc.plot(ax=ax, x='Date', y='Result_Value', ls='-', color='orange',label='Area of concern',linewidth=2.5)
    
    # add customizations
    plt.title(f'Monthly average of {coc}')
    plt.ylabel(f'{coc} (µg/L)')
    plt.xlabel('')
    plt.xticks(rotation = 90)
    plt.axhline(y=CULs[coc], color='red', linestyle='-', linewidth = 2, alpha = .75, label = 'Cleanup level')
    plt.legend()
    plt.savefig(f'../output/{coc}_trend.png')
```
